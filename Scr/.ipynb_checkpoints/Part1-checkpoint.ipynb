```python
#############################################################
# Recolectando Todos Los Ratings de todos los juegos en un data set
#############################################################
# Obteniendo ratings pagina por pagina para todos los juegos, pero en pedazos de 150 juegos
for numbers_of_chunks, chunks in df.groupby(np.arange(len(df))//150):
    # Creando un DF para aguantar todas las respuestas para este pedazo de juegos
    df_ratings = pd.DataFrame(columns=["gameid", "username", "rating"], index=range(int(chunks["nratings"].sum()+100000)))
    
    # Creando indices para escrituras para el ratings dataframe
    dfidx_start = 0
    dfidx = 0
    
    # Para este grupo de juegos, hacer llamadas hasta que Todas las paginas de cada juego hayan sido obtenidas
    #page_number = 1
    while len(chunks[chunks["nfullpages"] > 0]) > 0: 
        # Get a restricted DF with only still-active games (have ratings pages left)
        active_games = chunks[chunks["nfullpages"] > 0]
    
    # Set the next chunk of the DF "gameid" column using the list of game IDs
        id_list = []
        for df in active_games["gameid"]:
            id_list += [df]*100
        dfidx_end = dfidx_start + len(active_games)*100
        df_ratings.iloc[dfidx_start:dfidx_end, df_ratings.columns.get_loc("gameid")] = id_list
        
    # Make the request with the list of all game IDs that have ratings left
        id_strs = [str(gid) for gid in active_games["gameid"]]
        gameids = ",".join(id_strs)
        sleep(1.5)  # Keep the server happy
        r = request("http://www.boardgamegeek.com/xmlapi2/thing?id=%s&ratingcomments=1&page=1" % (gameids))        
        soup = BeautifulSoup(r.text, "xml")
        comments = soup("comment")
    """"    
    # Parse the response and assign it into the dataframe
        l1 = [0]*len(active_games)*100
        l2 = [0]*len(active_games)*100
        j = 0
        for comment in comments:
            l1[j] = comment["username"]
            l2[j] = float(comment["rating"])
            j += 1
        df_ratings.iloc[dfidx_start:dfidx_end, df_ratings.columns.get_loc("username")] = l1
        df_ratings.iloc[dfidx_start:dfidx_end, df_ratings.columns.get_loc("rating")] = l2
        
        chunks["nfullpages"] -= 1  # Decrement the number of FULL pages of each game id
        dfidx_start = dfidx_end     
        pagenum += 1  
        print("page number updated to %i" %(page_number,))
        
    # Strip off the empty rows
        df_ratings = df_ratings.dropna(how="all")
    # Write this batch of all FULL pages of ratings for this chunk of games to the DB
        df_ratings.to_csv(name="ratings", index=False)    
        print("Processed ratings for batch #%i of games." % (numbers_of_chunks))
        
    """"
    
```
